# Import library

```python
from bs4 import BeautifulSoup
```

```python
import requests
```

# settings scrapping website

```python
url ="https://www.flipkart.com/search?q=perfume&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
allow = requests.get(url)
```

```python
print(allow)
```

# testing scrapping allowability

```python
soup = BeautifulSoup(allow.text)
```

```python
soup
```

```python
cast = soup.select(".s1Q9rs")
```

```python
cast
```

```python
rinky=[]
```

```python
for item in cast:
    rinky.append(item.text)
```

```python
price=soup.select("._8VNy32 ._30jeq3")
```

```python
pinky=[]
```

```python
for item in price:
    pinky.append(item.text)
```

```python
pinky
```

```python
import pandas as pd
```
